```python
import pandas as pd

chunk_size = 500000  # rows per file

for i, chunk in enumerate(

    pd.read_json(
        "data_json.nosync/yelp_review.json",
        lines=True,
        chunksize=chunk_size
    ),
    start=1
):
    chunk.to_json(
        f"data_json.nosync/chunk_json/review_part_{i}.json",
        orient="records",
        lines=True,
        date_format = "iso"
    )
    print(f"Saved review_part_{i}.json")
```
