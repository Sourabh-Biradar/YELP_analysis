### YELP : Reviews & Sentiment Analysis

```python
# set-up s3

import boto3
from pathlib import Path
import os
from dotenv import load_dotenv
load_dotenv()

s3 = boto3.client(
    "s3",
    aws_access_key_id = os.getenv("AWS_ACCESS_KEY"),
    aws_secret_access_key = os.getenv("AWS_SECRET_ACCESS_KEY"),
    region_name = os.getenv("AWS_REGION")
)
```

```python
# upload reviews .json files to s3 bucket

bucket_name = os.getenv("BUCKET_NAME")

folder = Path("data_json.nosync/chunk_json")

for file in folder.glob("*.json"):
    s3.upload_file(
        Filename = str(file),
        Bucket = bucket_name,
        Key = f"yelp-reviews/{file.name}"
    )

    print(f"uploaded {file.name} successfully")
```

```python
# upload business .json file to s3 bucket

bucket_name = os.getenv("BUCKET_NAME")

file = "data_json.nosync/yelp_business.json"

s3.upload_file(
    Filename = str(file),
    Bucket = bucket_name,
    Key = f"yelp-business/yelp_business.json"
)

print(f"uploaded successfully")
```
